## Ejercicio 1
Somos una empresa que tiene varias tiendas de electrodomesticos. Necesitamos un programa para manejar las tiendas, ver las ventas que han tenido, cuántos empleados hay, etc... Para ello vamos a modelizar la tienda en una clase, que tendrá las siguientes características:
* Nombre clase: "Tienda"
* Atributos comunes:
    * Tipo: "Electrodomésticos"
    * Abierta: True
* Atributos propios de cada tienda:
    * Nombre - String
    * Dirección - String
    * Número de empleados - int
    * Ventas ultimos 3 meses - Lista de 3 numeros
* Método para calcular las ventas de todos los meses, que devuelva un numero con todas las ventas.
* Método que calcula la media de ventas de los ultimos meses, por empleado
* Método que devuelve en un string el nombre de la tienda, junto con su dirección.
* Método que obtiene las ventas del último mes.
* Método para dar una proyección de las ventas en caso de haber invertido X dinero en marketing en esa tienda. Siendo X el parámetro de entrada. Si X es menor de 1000, las ventas de los ultimos 3 meses hubiesen sido de (1.2 x ventas), si es mayor o igual de 1000, las ventas hubiesen sido de (1.5 x venta). El método reescribe el atributo de ventas en los últimos 3 meses, y además devuelve las ventas con el aumento.

Se pide:
1. Implementar la clase "Tienda"
2. Crear tres tiendas con datos inventados
3. Comprueba en al menos una de ellas, todo lo implementado en la clase tienda (sus atributos, media de ventas, ventas/empleado...)
3. Calcular las ventas del último mes de todas las tiendas. Para ello usa el bucle `for`

5. Seguro que ya has hecho este apartado... Documenta la clase :)

```python
class Tienda:
    """ Creamos una clase Tienda, para almacenar y calcular ventas de los ultimos tres meses.

        Atributos: 
        - tipo (String) y abierta (Booleano)
        
        El constructor llevará como argumentos distintivos:
        - Nombre (String), Dirección (String), Empleados (Integer), Ventas (lista)
        
        La clase consta de cinco métodos: 
        - ventas_meses, media_ventas_empleado, name_addres, ultimo_ventas_mes, inversion
    
    """

    tipo = 'Electrodomésticos'
    abierta = True
    
    def __init__(self, name: str, addres : str, employee : int, sales : list):
        self.name = name
        self.addres = addres
        self.employee = employee
        self.sales = sales
        
        
    def ventas_meses(self, sales : list) -> dict:
        """
        Calcula y retorna el valor de los últimos tres meses, ingresados en el constructor.
        
        Args:
            sales (list): --init--
            
        Returns:
            dict: con clave : mes y valor: sales
        """
        meses_ventas = {
            'mes 1': [self.sales[0]],
            'mes 2': [self.sales[1]],
            'mes 3': [self.sales[2]]
        }

        for mes in meses_ventas:
            for sales in meses_ventas[mes]:
                print(mes, ' = ', sales)


    def media_ventas_empleado(self, sales : list, employee: int) -> float:
        """
        Calcula y retorna la media de los últimos tres meses, utilizando datos ingresados previamente en el constructor.
        
        Args:
            sales (list): --init--
            employee (int): --init--
            
        Returns:
            float: de la media / entre empleados
        """
        
        total_ventas = sum(self.sales)
        media = total_ventas / len(self.sales)
        
        return media / self.employee
    

    def name_addres(self, nombre_tienda : str, addres : str) -> str:
        """
        Retorna el nombre y dirección de la tienda, utilizando datos previamente ingresados en el constructor.
            
        Args:
            nombre_tienda (str): --init--
            addres (str): --init--               
        """
        return f'"{self.name}" se encuentra en {self.addres}'

    def ultimo_ventas_mes(self, mes):   
        """
        Devuelve las ventas del ultimo mes
        
        Args: 
            mes (float): --init--
        """  
        return f'Las ventas del ultimo mes fueron de: {self.sales[-1]} euros'
    
    
    def inversion(self, sales : list, invers : float) -> float:
        """
        Calcula la aumento en las ventas con la inversión dada.
        
        Args:
            sales (list): --init--
            invers (float): inversión de marketing sobre nuestra tienda

        Returns:
            float: las ventas con la inversión ya aplicada.
        """
        
        if invers < 1000:
            venta = sum(self.sales) * 1.2
            # print(type(venta))
            self.sales = venta
            return f'La venta total de los meses, más el aumento de dicha inversión es de {self.sales} euros'     
        else:
            venta = sum(self.sales) * 1.5
            # print(type(venta))
            self.sales = venta
            return f'La venta total de los meses, más el aumento de dicha inversión es de {self.sales} euros'     
```

## Ejercicio 2
En este ejercicio vamos a implementar una clase *Perro* en Python. La clase tiene las siguientes características:
* Cosas que sabemos seguro que tiene un perro
    * Tiene 4 patas
    * 2 orejas
    * 2 ojos
    * Una velocidad de 0. Por defecto, el perro está parado
* Cuando se inicialice:
    * El perro será de una determinada raza
    * Por defecto tendrá pelo "Marrón", a no ser que se diga lo contrario.
    * Por defecto no tendrá dueño, a no ser que se diga lo contrario.
    
* Dispondrá también de un método llamado andar, que tiene un argumento de entrada (aumento_velocidad). Este valor se le sumará a la velocidad que ya llevaba el perro.
* Necesita otro método (parar), donde pondremos la velocidad a 0.
* Otro método llamado "ladrar", que tendrá un argumento de entrada, y la salida será el siguiente string: "GUAU!" + el argumento de entrada.


Se pide:
* Implementa la clase *Perro*
* Crea un objeto de tipo *Perro*, sin dueño
* Comprueba que están bien todos sus atributos
* Prueba que ande, y comprueba su velocidad
* Páralo
* Documenta la clase *Perro*

```python
class Perro:
    """
    Creamos una clase Perro, que actuará con ciertas funciones del propio animal
    Atributos:
    - patas, orejas, ojos, velocidad
    
    Tendrá una función constructora:
    - raza, color, duenio
    
    Constará de tres métodos:
    - andar, parar, ladrar.
    """
    patas = 4
    orejas = 2
    ojos = 2
    velocidad = False
    
    def __init__(self, raza : str, color='marron', duenio=False):
        """Constructor de la clase perro."""
        self.raza = raza
        self.color = color
        self.duenio = duenio
        
    def andar(self, velocidad: int):
        actual_movimiento = int(self.velocidad) + velocidad
        
        return f'La velocidad de su perro es de: {actual_movimiento}'
    
    def parar(self):
        self.velocidad = False
        return f'El perro esta quieto : velocidad = {int(self.velocidad)}'
    
    def ladrar(self, input: str):
        return 'GUAU!! ' + input
    
    
    
```

```python
perro_1 = Perro('chihuahua','blanco',duenio=True)

print('Patas:', perro_1.patas,'Orejas:', perro_1.orejas,'Ojos:', perro_1.ojos,'Speed:', perro_1.velocidad)

print(perro_1.andar(10))
print(perro_1.parar())

print(perro_1.ladrar('corre, gil'))
```
