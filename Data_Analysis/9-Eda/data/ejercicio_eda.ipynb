## Cardinalidad

```python
df = pd.DataFrame({'columna1': [1,2,2,3], 'Columna2': [4,5,6,7]})
df
```

```python
def funcion(df : pd.DataFrame) -> pd.DataFrame:
    unicos = list()
    for i in df.columns:
        cardinalidad = df[i].nunique()
        porcentaje_cardinalidad = df[i].nunique() / len(df)
        tipo_de_dato = df[i].dtypes
        unicos.append(df[i].unique())
        
        tipo = input(f"'cardinalidad': {cardinalidad}\n % cardinalidad: {porcentaje_cardinalidad}\n tipo de datos: {tipo_de_dato}: ")
        
        
    df = pd.DataFrame({'cardinalidad': cardinalidad, '% cardinalidad': porcentaje_cardinalidad, 'tipo de datos': tipo_de_dato, 'valores unicos': unicos, 'tipo de variable': tipo})
    return df

#Me falta el nombre de la columnna
```
